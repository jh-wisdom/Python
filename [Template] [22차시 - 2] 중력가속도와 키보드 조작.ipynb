```python
GlowScript 3.1 VPython

# 공이 밑바닥의 좌우 끝 범위에 닿으면 튕기고 아니면 떨어지기

ball = sphere(pos=vector(0,10,0), radius=1, color=color.red)
floor = box(pos=vector(0,0,0), size = vector(10,0.5,10), color=color.green)

ball.velocity=vector(0,0,0)
dt=0.01
dx=0.05
g =-9.8 
t=0

while True:
  
  rate(100)
  ball.pos=ball.pos+ball.velocity*dt
  k = keysdown()
  if 'left' in k : floor.pos.x-=dx
  if 'right' in k : floor.pos.x+=dx
  
  
  if (ball.pos.y<=floor.pos.y+1.25): # 첫 번째 질문! 공이 밑바닥 상반부에 닿았는가?
     if (ball.pos.x+>=(floor.pos.x-) and ball.pos.x-<=(floor.pos.x+)): # 두 번째 질문! 공이 밑바닥의 왼쪽 끝과 오른쪽 끝에 닿았는가 
        ball.velocity.y = -ball.velocity.y
  else:
   ball.velocity.y=ball.velocity.y+g*dt #가속도를 속도에 반영해주기
  t=t+dt
  
```
