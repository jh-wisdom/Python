### 여러 개의 물체를 만들 때 애니메이션 효과 주기!

* 물체를 만들 때 한번에 만들도록 하는게 아니라
* 물체를 만들고 다음 물체를 만들 때의 간격을 지정해주면 애니메이션 효과를 지정할 수 있다.
* rate(숫자)를 입력하면 지연을 줄 수 있다.
* 간격은 1/숫자로 계산된다.
* 예를 들어 rate(1)을 한다면 1초 간격
* rate(50)을 한다면 1/50 -> 0.02초 간격을 줄 수 있다.

```python
GlowScript 3.1 VPython

#규칙을 살려서 5개의 공을 일렬로 세워봅시다. (홀수 공은 빨간색, 짝수 공은 파란색)

for i in range(): 
    rate()
    if i %   == 0:
        ball = sphere(pos=vector(0,0+(),0), radius=10, color=color.)
    else :
        ball = sphere(pos=vector(0,0+(),0), radius=10, color=color.)
```
