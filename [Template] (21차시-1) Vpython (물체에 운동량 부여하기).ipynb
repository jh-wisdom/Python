### 가속도 운동

- 등속 운동이 동일한 속도로 운동하는 운동이었다면
- 가속도 운동은 동일한 속도가 아닌 속도가 변하는 운동이다.
- 여기서는 중력 가속도를 적용하여 가속도 운동하는 물체를 만들어보자!

```python
GlowScript 3.1 VPython

ball = sphere(pos=vector(0,10,0), radius=3, color=color.red)

ball.velocity=vector(0,0,0)
dt=0.01 #시간의 변화 0.01로 바꿈 
g =-9.8 #중력가속도
t=0

while True: 
  rate(100)
  ball.pos=ball.pos+ball.velocity*dt
  ball.velocity.y=ball.velocity.y+g*dt #가속도를 속도에 반영해주기
  t=t+dt
```
