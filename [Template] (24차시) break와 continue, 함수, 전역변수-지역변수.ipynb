### while(1)과 break, continue

- 지난 시간에 우리는 while(1)을 통해 공의 움직임을 무한대로 지정하여 실행해보았습니다.
- 그런데 무한대로 실행되는 프로그램이 의미가 있을까요?
- 설령 아무리 완벽한 프로그램이라 하더라도 시간이 지나면 그 프로그램을 변한 상황에 맞게 유지보수해야 할 필요가 있습니다.
- 그리고 세상에 무한하게 돌아가야 할 필요성이 있는 프로그램은 얼마 없죠. 따라서 while(1)을 이용한 무한 반복문 또는 무한이 아니더라도 큰 반복을 하는 경우는 상황에 따라 프로그램을 종료하거나, 다시 반복 조건을 체크해주어야 할 필요가 있습니다.
- 여기서는 자신을 제외한 (파이선에서는 자신이 있는 들여쓰기 (안쪽으로 들어간)를 제외한 반복문) 반복문을 종료해주는 break
- 마찬가지로 자신을 제외한 반복문으로 돌아가게 하는 continue에 대해 알아보겠습니다.

### 이렇듯 함수를 사용하면

- 함수와 실행 부분을 분리할 수 있어 프로그램의 영역을 구별하기 쉽다.
- 실행 파일과 함수 파일을 따로 만들어 저장하면 영역, 기능별로 알아보기도 쉽다.
- 실행 파일이 해킹되더라도 함수를 별도 파일로 만들어 암호화해 저장하면 해킹의 위험도 줄일 수 있다.
- 또 한 명의 사람이 프로그램을 만드는 것이 아니라 여러 사람이 프로그램을 만들 수 있다. (A 프로그래머는 1 함수, B 프로그래머는 2 함수... 만든 후 합치기)
- 또 프로그램의 기능을 변경해야 할 일이 생겼을 때 각각의 함수 부분만 수정하면 실행 파일(함수를 사용한다 호출하는 부분)을 변경할 필요도 없다.
- 이런 함수 사용도 한계가 있는데 그것을 해결한 개념이 객체지향 프로그래밍(OOP)이다.
- 하지만 OOP는 여기서는 안다룬다 ^^; (여러분 모두가 전문 프로그래머가 될 필요도 없고 이유도 없다~)
