## 7. 반복문 
```
for 변수 in range(초기값, 최종값, [증감]):
    실행문
```
- 초기값 ~ (최종값-1)까지 증 반복 (+일때 미만)
- 초기값 ~ (최종값+1)까지 감 반복 (-일때 초과)

````python
for i in range(,,):
  print(i)
````

````python
#9부터 1까지 출력하기 
for i in range(,,):
  print(i)
````

````python
#(문제) 1,3,5,7,9 출력하기 (\n없애기)
for i in range(1,,):
  print(i,end='')
````

````python
#(문제) 10,8,6,4,2,0 출력하기 (\n : 줄바꿈 없애기)
for i in range(10,,):
  print(i,end="")
````

````python
#숫자(5) 삼각형 출력하기

for i in range(1,6):
  for j in range(1,i+1):
    print(j,end=' ')
  print()
````
